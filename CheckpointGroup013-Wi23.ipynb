# Evaluation Metrics

We wish to use Confusion Matrix, specifically the values of Recall in order to limit false negatives. Recall is defined as “TruePositives / (TruePositives + FalseNegatives).” The reason why we are using Recall is due to the fact that it can be extremely harmful if someone has a heart disease but the model predicts that they don't – a false negative – which is measured by Recall. Specifically, Recall will be a value between 0 and 1, with 0 being the worst and 1 being perfect Recall. This model is able to tolerate false positives (Measured by Precision) because that would mean that a patient not diagnosed with heart disease is tested positive, all that leads to for the patient is further medical analysis. If we evaluate our model with the F1 metric, which equally gives weight to Recall and Precision, it would put some weight away from prioritizing the maximization of the Recall value. However, we will still be also looking at other metric that involve FalseNegatives such as False Negative Rate, False Omission Rate, and F1 score.

```python
print("Duplicated data points:", df.duplicated().sum())
print("Current shape:", df.shape)
df = df.drop_duplicates()
print("Without duplicates:", df.shape)
df = df.dropna()
print("Without data points with missing info:", df.shape, "\n")
print(df.describe())

print("\nSleep counts:\n")
print(df.SleepTime.value_counts())
```

EDA

Like we stated in the data portion, we removed the features that we believe to be less critical based on the following EDA graphs that display our data.

```python
d = df.copy()
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
for col in d.columns:
    if d[col].dtype == object:
        d[col] = label.fit_transform(d[col])

plt.figure(figsize=(15,15))
cor = d.corr()
sns.heatmap(cor, annot=True, cmap=plt.cm.Blues, fmt='.2f')
plt.show()
```

```python
#print(d.BMI.describe())

prange = np.arange(10, 100, 10)
nohd = d.loc[d.HeartDisease == 0]
hd = d.loc[d.HeartDisease == 1]

binnedNOHD = nohd.groupby(pd.cut(nohd.BMI, prange)).count().HeartDisease.rename('Count').reset_index()
binnedNOHD.Count = binnedNOHD.Count / binnedNOHD.Count.sum()

binnedHD = hd.groupby(pd.cut(hd.BMI, prange)).count().HeartDisease.rename('Count').reset_index()
binnedHD.Count = binnedHD.Count / binnedHD.Count.sum()

binnedHD["HeartDisease"]=1
binnedNOHD["HeartDisease"]=0
c_df = pd.concat([binnedNOHD, binnedHD])
print(c_df)

c_df.HeartDisease.replace({1:"Heart Disease", 0:"No Heart Disease"}, inplace=True)
ax = sns.barplot(x='HeartDisease', y='Count', hue='BMI', data=c_df, palette=sns.color_palette("Set2"))
ax.set_title("BMI and Heart Disease",fontsize=24)
ax.set_xlabel('Heart Disease',fontsize=16)
ax.set_ylabel('Proportion of Subjects',fontsize=16)

plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0)
sns.set(rc={'figure.figsize':(8,6)})
plt.show()
```

In the BMI graph we saw that there appeared to be a slight right skew for both the non heart disease and heart disease observations. Since there appears to be some sort of behavior present we decided to keep this one. Even though the distributions of the two groups appear rather similar, as this is a physical health metric and thus a bit more critical in heart disease predictions. We believe that it will have some influence on our prediction of heart disease.

```python
c_df = d.groupby(['HeartDisease'])['Smoking'].value_counts(normalize=True).rename('smoke_count').reset_index()

c_df.Smoking.replace({1:"Smoker", 0:"Non-smoker"}, inplace=True)
c_df.HeartDisease.replace({1:"Heart Disease", 0:"No Heart Disease"}, inplace=True)

print(c_df)

smoking_hd = sns.barplot(x='HeartDisease', y='smoke_count', hue='Smoking', data=c_df, palette=sns.color_palette("hls", 9))
smoking_hd.set_title("Smoking and Heart Disease",fontsize=24)
smoking_hd.set_xlabel('Heart Disease',fontsize=16)
smoking_hd.set_ylabel('Proportion of Subjects',fontsize=16)

plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0)
sns.set(rc={'figure.figsize':(8,6)})
plt.show()
```

```python
# Building a model using KNeighborsClassifier 
def evaluate_model(model, x_test, y_test):
    from sklearn import metrics

    # Predict Test Data 
    y_pred = model.predict(x_test)

    # Calculate accuracy, precision, recall, f1-score, and kappa score
    acc = metrics.accuracy_score(y_test, y_pred)
    rec = metrics.recall_score(y_test, y_pred)
    f1 = metrics.f1_score(y_test, y_pred)

    # Calculate area under curve (AUC)
    y_pred_proba = model.predict_proba(x_test)[::,1]

    # Display confussion matrix
    cm = metrics.confusion_matrix(y_test, y_pred)

    return {'acc': acc, 'rec': rec, 'f1': f1, 'cm': cm}
knn = KNeighborsClassifier(n_neighbors = 5)

knn.fit(X_train, y_train)


# Evaluate Model
knn_eval = evaluate_model(knn, X_test, y_test)

# Print result
print('Accuracy:', knn_eval['acc'])
print('Recall:', knn_eval['rec'])
print('F1 Score:', knn_eval['f1'])
print('Confusion Matrix:\n', knn_eval['cm'])
```

```python
####### LOGISTIC REGRESSION #######

# Input hyperparameters in the grid

scaler = StandardScaler()

param_grid = {
    'logistic__C':[0.0005, 0.01, 0.5, 10, 5e2],
    'logistic__solver':['saga','liblinear'],
    'logistic__penalty':['l1', 'l2']
}
logistic = LogisticRegression()
# Run the pipeline
kf = KFold(n_splits=5)
imp = SimpleImputer(missing_values=np.nan, strategy='mean')
pipe = Pipeline(steps=[("scaler", scaler), ("imputer", imp), ("logistic", logistic)])

# Now, we want to use GridSearchCV
gscv = GridSearchCV(pipe, param_grid, cv = kf, scoring = 'recall')
gscv.fit(X_train, y_train)
gscv_best = gscv.best_estimator_
    
uni_labels = np.unique(y_test)
y_pred = gscv_best.predict(X_test)
y_pred_train = gscv_best.predict(X_train)
recall = recall_score(y_test, y_pred, labels = uni_labels)
f1 = f1_score(y_test, y_pred, labels = uni_labels)
print('Recall', recall)
print('F1', f1)

# Make graphs of the confusion_matrix
cm = confusion_matrix(y_test, y_pred, labels = uni_labels)
disp = ConfusionMatrixDisplay(cm)
disp.plot()
plt.show()

FN = cm.sum(axis = 1) - np.diag(cm)
FP = cm.sum(axis = 0) - np.diag(cm)
TP = np.diag(cm)
TN = cm.sum() - (FP + FN + TP)

FN = FN.astype(float)[1]
FP = FP.astype(float)[1]
TP = TP.astype(float)[1]
TN = TN.astype(float)[1]

FNR = FN / (TP + TN)
FOR = FN / (TN + FN)
print('False Negative Rate:', FNR)
print('False Omission Rate:', FOR)

# Roc_auc
fpr, tpr, thresholds = roc_curve(y_test, y_pred) 
roc_auc = auc(fpr, tpr)
disp = RocCurveDisplay(fpr=fpr, tpr=tpr, roc_auc=roc_auc)
disp.plot()
plt.show()

# Accuracy scores for training and test sets
print("Training accuracy", accuracy_score(y_train, y_pred_train))
print("Testing accuracy", accuracy_score(y_test, y_pred))
print("Classifcation Report", classification_report(y_test, y_pred))
print("Best Params", gscv.best_params_)
print("Best score", gscv.best_score_)
```
